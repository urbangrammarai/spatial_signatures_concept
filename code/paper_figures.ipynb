# Maps for the paper

This notebook creates maps used in the paper.

```python
import geopandas as gpd

import urbangrammar_graphics as ugg
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
from matplotlib.lines import Line2D
import contextily
from shapely.geometry import Point
```

```python
sns.set(context="paper", style="ticks", rc={'patch.force_edgecolor': False})
```

```python
folder = "../../urbangrammar_samba/sp_sig_concept/barcelona/"
```

```python
barcelona = gpd.read_parquet(folder + "signatures.pq")
```

```python
enc = gpd.read_parquet(folder + "enclosures.pq")
```

```python
cmap = ugg.get_colormap(barcelona.cluster.nunique(), randomize=False)
```

```python
cols = cmap.colors
symbology = {0: cols[17],
             1: cols[4],
             2: cols[8],
             3: cols[6],
             4: cols[14],
             5: cols[7],
             6: cols[9],
             7: cols[13],
             8: cols[11],
             9: cols[12],
             10: cols[10],
             11: cols[15],
             12: cols[5],
             13: cols[3],
             14: cols[16],
             15: cols[0]
            }
```

```python
df = barcelona.set_crs(enc.crs).to_crs(3857)
```

```python
token = ""
```

```python
ax = df.plot(color=df['cluster'].map(symbology), figsize=(14, 14), zorder=1, linewidth=.3, edgecolor='w', alpha=1)
contextily.add_basemap(ax, crs=df.crs, source=ugg.get_tiles('roads', token), zorder=2, alpha=.3)
contextily.add_basemap(ax, crs=df.crs, source=ugg.get_tiles('background', token), zorder=-1, alpha=1)
ax.set_axis_off()

scalebar = ScaleBar(dx=1,
                    color=ugg.COLORS[0],
                    location='lower right',
                    height_fraction=0.002,
                    pad=.5,
                    frameon=False,
                    fixed_value=10_000,
                    units="m",
                    )
ax.add_artist(scalebar)

# ugg.north_arrow(plt.gcf(), ax, 0, size=.05, linewidth=1, color=ugg.COLORS[0], loc="upper left", pad=.002, alpha=.9)

custom_points = [Line2D([0], [0], marker="o", linestyle="none", markersize=10, color=color) for color in symbology.values()]
leg_points = ax.legend(custom_points, symbology.keys(), bbox_to_anchor=(1.05, 1), loc='upper left', frameon=True)
ax.add_artist(leg_points)
plt.savefig("../paper/figures/barcelona.png", dpi=300, bbox_inches="tight")
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres')).dissolve().to_crs("+proj=wintri")
```

```python
gdf = gpd.GeoDataFrame({"city": ["Barcelona", "Medellin", 'Dar es Salaam', 'Singapore', 'Houston']},
                       geometry=[Point(1.41255962598549, 42.110661849474675),
                                Point(-75.6228907936514, 6.324592026693382),
                                Point(39.21329108932159, -6.8733488834917935),
                                Point(103.9235428728682,1.276390943209925),
                                Point(-95.32682393205496, 29.83517835513806)],
                       crs=4326
).to_crs("+proj=wintri")
```

```python
ax = world.boundary.plot(color=ugg.COLORS[0], figsize=(15, 15), linewidth=1)
gdf.plot(ax=ax, color=ugg.COLORS[1:], markersize=200, alpha=.8, zorder=2)
ax.set_axis_off()
ax.annotate("",
            xy=(gdf.geometry.x[0], gdf.geometry.y[0]), xycoords='data',
            xytext=(gdf.geometry.x[0] + 5000000, gdf.geometry.y[0] + 6000000), textcoords='data',
            size=20, va="center", ha="center",
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3,rad=0.2", color=ugg.COLORS[1]),
            color=ugg.COLORS[0],
            fontname = 'Times',
            )
ax.annotate("",
            xy=(gdf.geometry.x[1], gdf.geometry.y[1]), xycoords='data',
            xytext=(gdf.geometry.x[1] - 2000000, gdf.geometry.y[1] - 2000000), textcoords='data',
            size=20, va="center", ha="center",
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3,rad=0.2", color=ugg.COLORS[2]),
            )
ax.annotate("",
            xy=(gdf.geometry.x[2], gdf.geometry.y[2]), xycoords='data',
            xytext=(gdf.geometry.x[2] + 2000000, gdf.geometry.y[2] - 2000000), textcoords='data',
            size=20, va="center", ha="center",
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3,rad=0.2", color=ugg.COLORS[3]),
            color=ugg.COLORS[3],
            fontname = 'Times',
            )
ax.annotate("",
            xy=(gdf.geometry.x[3], gdf.geometry.y[3]), xycoords='data',
            xytext=(gdf.geometry.x[3] + 4000000, gdf.geometry.y[3] + 2000000), textcoords='data',
            size=20, va="center", ha="center",
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3,rad=0.2", color=ugg.COLORS[4]),
            color=ugg.COLORS[4],
            fontname = 'Times',
            )
ax.annotate("",
            xy=(gdf.geometry.x[4], gdf.geometry.y[4]), xycoords='data',
            xytext=(gdf.geometry.x[4] - 2000000, gdf.geometry.y[4] - 2000000), textcoords='data',
            size=20, va="center", ha="center",
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3,rad=-0.2", color=ugg.COLORS[5]),
            color=ugg.COLORS[5],
            fontname = 'Times',
            )
# plt.savefig("/Users/martin/Git/ug_talks/fig/sp_sig/examples_map.svg", bbox_inches="tight")
```
